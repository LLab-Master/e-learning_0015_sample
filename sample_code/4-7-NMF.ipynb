```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Iris 
from sklearn.datasets import load_iris
import pandas as pd
from pandas import plotting

iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

print(iris_df)

plotting.scatter_matrix(iris_df, figsize=(10, 10), c=iris.target)
plt.show()
```

```python
# NMF 
from sklearn.decomposition import NMF
nmf = NMF(n_components=2, max_iter=600)
feature = nmf.fit_transform(iris_df)

# 
plot0 = feature[iris.target == 0]
plot1 = feature[iris.target == 1]
plot2 = feature[iris.target == 2]

plt.scatter(plot0[:,0], plot0[:,1], c='r', label=iris.target_names[0])
plt.scatter(plot1[:,0], plot1[:,1], c='g', label=iris.target_names[1])
plt.scatter(plot2[:,0], plot2[:,1], c='b', label=iris.target_names[2])
plt.xlabel('new_feature1')
plt.ylabel('new_feature2')
plt.legend()
```
