```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from sklearn import preprocessing

## 初期データ
x = np.array([39, 49,  7, 22, 55, 
     64, 36, 40, 55, 65, 
     17, 26, 58, 34, 77, 
     63, 89, 50, 42, 78])
print('平均=', x.mean(), '  標準偏差=', x.std())
plt.hist(x, rwidth=0.9)
plt.show()

# 標準化
x_trans = preprocessing.scale(x)
print('平均=', x_trans.mean(), '  標準偏差=', x_trans.std())
plt.hist(x_trans, rwidth=0.9)
plt.show()

```
