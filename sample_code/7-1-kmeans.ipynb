```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

X = np.loadtxt('claster5.csv', delimiter=',')

plt.scatter(X[:,0], X[:,1])
plt.show()


from sklearn.cluster import KMeans

# K-means 
km = KMeans(n_clusters=5)

# 学習 & 予測
labels = km.fit_predict(X)

# 図示
plt.scatter(X[:,0], X[:,1], c=labels)
plt.show()
```
